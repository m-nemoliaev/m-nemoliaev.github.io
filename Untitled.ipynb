```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movie_dataset.csv")
```

```python
df.info()
```

```python
df["budget_mil"] = df["budget"] / 1000000
df["ratio_bp"] = df["budget_mil"] / df["popularity"]
df["ratio_pb"] = df["popularity"] / df["budget_mil"]
```

```python
count = 0
for i in range(len(df["budget_mil"])):
    if df["budget_mil"][i] >= 150:
        count += 1
print(count/len(df["budget_mil"])*100 )
```

```python
count = 0
for i in range(len(df["popularity"])):
    if df["popularity"][i] <= 12.9216:
        count += 1
print(count/len(df["popularity"])*100)
```

```python
sns.scatterplot(data=df, x="budget_mil", y="popularity")
```

```python
count = 0
for i in range(len(df["vote_average"])):
    if df["vote_average"][i] >= 6.2:
        count += 1
print(count/len(df["vote_average"])*100)
```
